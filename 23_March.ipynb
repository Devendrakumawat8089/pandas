```python
import plotly.graph_objects as go 
import plotly.io as pio 
pio.templates.default='plotly_white'
```

```python
print(df.isnull().sum())
```

```python
# check Persentage of attribute by department 
```

```python
# filter the data to show only "Yes" Values in the "Attrition" Column 
```

```python
attr_diif = df[df['Attrition'] == 'Yes']
```

```python
attr_diif
```

```python
# calculate the attrition by department 
```

```python
attr_by_dpt = attr_diif.groupby(['Department']).size().reset_index(name='count')
```

```python
attr_by_dpt
```

```python
# Create a Donut chart 
```

```python
fig = go.Figure(data=[go.Pie(
    labels = attr_by_dpt['Department'],
    values = attr_by_dpt['count'],
    hole = 0.4,
    marker = dict(colors=['red','green']),
    textposition = 'inside'
)])
```

```python
# Update the Layout 
fig.update_layout(title = "Attrition by Department")
# fig.show() #Error ?
```
